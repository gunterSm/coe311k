```python
#Manual LSR
import matplotlib.pyplot as plt
import numpy as np 
#My equation is y= 3.5x+8
k = 3.5
l = 8
X = np.arange(0,1, 0.01)


Y = k*X + l

#The noise
noise = 3*(np.random.rand(len(X))- 0.5)
#The equation with some noise
y = Y - noise
#Matrix A
A = np.vstack([X, np.ones(len(X))]).T

#Calculating coefficients
coeff = np.linalg.inv(A.T @A) @ A.T @y
m, c = coeff[0], coeff[1]

print("The slope is:", m)
print("The y intercept is:", c)


plt.scatter(X,y, marker = '^', color = 'y')
plt.plot(X, m*X+c, linestyle= 'solid', color = 'k')
plt.title("Best Fitting Line")
plt.show()

#Residuals
y_predicted = m*X +c
res = y - y_predicted

print("This is the largest residual:", np.max(res))
print("This is the smallest residual:", np.min(res))
```
