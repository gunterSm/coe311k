```python
#Jacobian method with Tolerance of 10^-3
import time
import numpy as np
import matplotlib.pyplot as plt
def jacobi(A, b, x0, tol, max_iterations):
    Error_plot = []
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        plot1 = np.linalg.norm(x_new - x, ord=np.inf)
        Error_plot.append(plot1)
# Check for convergence
        if plot1 < tol:
            return x_new, k, Error_plot
        x = x_new
    return x, max_iterations, Error_plot

#Matrices 
A = np.array([[3, -0.1, -0.2],
              [0.1, 7, -0.3],
              [0.3, -0.2, 10]])

b = np.array([7.85, -19.3, 71.4])

x0 = np.zeros_like(b)

tol = 1e-3
max_iterations = 100
start_time = time.perf_counter()

solution, iterations, Error_plot = jacobi(A, b, x0, tol, max_iterations)

end_time = time.perf_counter()
elapsed_time = end_time - start_time

print("The elapsed time was:", elapsed_time, "seconds")
print(f"Solution Vector: {solution}")
print(f"Iterations: {iterations}")
print(f"Error Points: {Error_plot}")



```

```python
#Gaussian Model with Tolerance of 10^-3
import time
import numpy as np
def gauss_seidel(A, b, x0, tol, max_iterations):
    Gauss = []
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        p = np.linalg.norm(x_new - x, ord=np.inf)
        Gauss.append(p)    
# Check for convergence
        if p < tol:
           return x_new, k, Gauss
        x = x_new
    return x, max_iterations, Gauss
#Example Usage
A = np.array([[3, -0.1, -0.2],
              [0.1, 7, -0.3],
              [0.3, -0.2, 10]])

b = np.array([7.85, -19.3, 71.4])

x0 = np.zeros_like(b)
tol = 1e-3
max_iterations = 100
start_time = time.perf_counter()

solution, iterations, Gauss = gauss_seidel(A, b, x0, tol, max_iterations)

end_time = time.perf_counter()
elapsed_time = end_time - start_time

print("The elapsed time was:", elapsed_time, "seconds")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Gaussian for Tolerance of 10^-3: {Gauss}")
print("\n")

#Gaussian Model with Tolerance of 10^-6
import time
import numpy as np
def gauss_seidel(A, b, x0, tol, max_iterations):
    Gauss1 = []
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        p = np.linalg.norm(x_new - x, ord=np.inf)
        Gauss1.append(p)    
# Check for convergence
        if p < tol:
           return x_new, k, Gauss1
        x = x_new
    return x, max_iterations, Gauss1
#Example Usage
A = np.array([[3, -0.1, -0.2],
              [0.1, 7, -0.3],
              [0.3, -0.2, 10]])

b = np.array([7.85, -19.3, 71.4])

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
start_time = time.perf_counter()

solution, iterations, Gauss1 = gauss_seidel(A, b, x0, tol, max_iterations)

end_time = time.perf_counter()
elapsed_time = end_time - start_time

print("The elapsed time was:", elapsed_time, "seconds")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Gaussian for Tolerance of 10^-6: {Gauss1}")
print("\n")

#Gaussian Model with Tolerance of 10^-9
import time
import numpy as np
def gauss_seidel(A, b, x0, tol, max_iterations):
    Gauss2 = []
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        p = np.linalg.norm(x_new - x, ord=np.inf)
        Gauss2.append(p)    
# Check for convergence
        if p < tol:
           return x_new, k, Gauss2
        x = x_new
    return x, max_iterations, Gauss2
#Example Usage
A = np.array([[3, -0.1, -0.2],
              [0.1, 7, -0.3],
              [0.3, -0.2, 10]])

b = np.array([7.85, -19.3, 71.4])

x0 = np.zeros_like(b)
tol = 1e-9
max_iterations = 100
start_time = time.perf_counter()

solution, iterations, Gauss2 = gauss_seidel(A, b, x0, tol, max_iterations)

end_time = time.perf_counter()
elapsed_time = end_time - start_time

print("The elapsed time was:", elapsed_time, "seconds")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Gaussian for Tolerance of 10^-9: {Gauss2}")
print("\n")
```
