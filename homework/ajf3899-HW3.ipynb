```python
import numpy as np 
import math
import matplotlib.pyplot as plt

def TaylorCalculation(x):
    p= 0
    L = []
    for j in range(4):
        function = ((-1**j))*(x**(2*j+1))/math.factorial(2*j+1) #You have no idea how much trouble I had with parenthesis
        p = p +function
        L.append(p)
    return L

x_range = np.arange(0,10,0.1)  #Wanted to use linspace, but apparently it cant do decimal step sizes
sin_values = np.sin(x_range) #true values of the function of sin within the specified range

List1, List2, List3, List4 = [], [], [], []
error_one, error_two, error_three, error_fourth = [], [], [], []
 
for i in range(len(x_range)):#I searched up for the len because it gives you the values of all indexes values for each x looped
    x = x_range[i]
    first, second, third, fourth = TaylorCalculation(x) 
    List1.append(first)
    List2.append(second)
    List3.append(third)
    List4.append(fourth) 
    
List1 = np.array(List1)
List2 = np.array(List2)
List3 = np.array(List3)
List4 = np.array(List4)
    
error_one = np.abs(sin_values - List1)
error_two = np.abs(sin_values - List2)
error_three = np.abs(sin_values - List3)
error_fourth = np.abs(sin_values - List4)
    

plt.plot(x_range, error_one, label= "List 1")
plt.plot(x_range, error_two, label= "List 2")
plt.plot(x_range, error_three, label = "List3")
plt.plot(x_range, error_fourth, label = "List4")
plt.xlabel("Range")
plt.ylabel("Truncation Error")
plt.legend()
plt.grid()
plt.show()

#From the graph, it appears that the error grows exponentially for every iteration you give to the function, the truncation error will increase
#So, as you go further, the truncation error begins to grow more exponentially because you are farther away from the 0.
#Also, I hope this is right....
```
