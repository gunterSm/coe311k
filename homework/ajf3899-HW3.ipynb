```python
import numpy as np 
import math
import matplotlib.pyplot as plt

def TaylorCalculation(x):
    p= 0
    for j in range(4):
        function = ((-1**j))*(x**(2*j+1))/math.factorial(2*j+1)
        p = p +function
    return p

x_range = np.arange(0,10,0.1)  #Wanted to use linspace, but apparently it cant do step size

sin_values = [np.sin(x) for x in x_range] #true values of the function of sin within the specified range
approximated_sin_values = [TaylorCalculation(x) for x in x_range] #Store values of the function into a list in the range from 0 to 10 with step size of 0.1

absolute_error = []  
for i in range(len(x_range)):     #I searched up for the len because it gives you the values of all indexes values for each x looped
    error = np.abs(sin_values[i] - approximated_sin_values[i])
    absolute_error.append(error)

plt.plot(x_range, absolute_error,'g')
plt.xlabel("Range")
plt.ylabel("Truncation Error")
plt.grid()
plt.show()

#From the graph, it appears that the error grows exponentially for every iteration you give to the function
```

```python
def BigTaylor(x):
    p= 0
    for j in range(1):
        function = ((-1**j))*(x**(2*j+1))/math.factorial(2*j+1)
        p = p + function
    return p

x_range = np.arange(0,10,0.1)  
sin_values = [np.sin(x) for x in x_range] 
approximated_sin_values = [BigTaylor(x) for x in x_range]

absolute_error = []  
for i in range(len(x_range)):     
    error = np.abs(sin_values[i] - approximated_sin_values[i])
    absolute_error.append(error)

plt.plot(x_range, absolute_error, 'm')
plt.xlabel("Range")
plt.ylabel("Truncation Error")
plt.grid()
plt.show()

#Here, if you used 4000 terms instead of 4, this causes an overflow error because it can no longer convert the decimals. 
#If you use 1 instead the graph start looking weird, almost like an s shape.Basically the truncation has periods where it increases exponentially and then saddles. 
```
