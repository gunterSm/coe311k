```python
import random
import matplotlib.pyplot as plt
import numpy as np

def Montecarlo():
    Iterations = []
    Estimate = []
    for i in range(10, 10000,20):
        dots = 0
        for j in range(i):       
            x = random.uniform(-1, 1)
            y = random.uniform(-1, 1)

            if((x*x + y*y) < 1):
                dots += 1
            
            pi_estimation= 4.0 *(dots/ i)
        
        Iterations.append(i)
        Estimate.append(pi_estimation)
    plt.plot(Iterations, Estimate, 'g')
    plt.axhline(np.pi, linestyle="-")
    plt.xlabel("Number of Samples")
    plt.ylabel("Pi Estimate")
    plt.title("Estimation of Pi")
    plt.show()  
Montecarlo()    


   
```
