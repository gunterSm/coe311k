```python
import random
import matplotlib.pyplot as plt
import numpy as np

def Montecarlo():
    Iterations = []
    Error = []
    for i in range(10, 10000,20):
        dots = 0
        for j in range(i):       
            x = random.uniform(-1, 1)
            y = random.uniform(-1, 1)

            if((x*x + y*y) < 1):
                dots += 1
            
            pi_estimation= 4.0 *(dots/ i)
            error = abs(np.pi - pi_estimation) 
        
        Iterations.append(i)
        Error.append(error)
    plt.plot(Iterations, Error, 'g')
    plt.xlabel("Number of Samples")
    plt.ylabel("Pi Estimate")
    plt.title("Estimation of Pi")
    plt.show()  
Montecarlo()    


   
```
