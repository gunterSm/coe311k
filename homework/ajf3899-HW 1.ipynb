```python
import random
import matplotlib.pyplot as plt
import numpy as np

def Montecarlo():
    Samples = [] #I added two lists to store the values for each iterations
    Error = [] 
    for i in range(10, 10000,20): 
        dots = 0           #added two counters, one for the range of samples, and the other to store and count the value of the dots
        for j in range(i):       
            x = random.uniform(-1, 1)
            y = random.uniform(-1, 1) #Regular constraints within the circle 

            if((x*x + y*y) < 1):
                dots += 1         #Adding a dot for every time there is a value inside the function 
            
            pi_estimation= 4.0 *(dots/ i)
            error = abs(np.pi - pi_estimation)  #Im subtracting to get the absolute error 
        
        Samples.append(i)
        Error.append(error) #Adding every iteration value to its respective number of samples and error
    plt.plot(Samples, Error, 'g') #I like green
    plt.xlabel("Number of Samples")#I searched the labels' command on google, sorry :3
    plt.ylabel("Error")
    plt.title("Estimation of Pi")
    plt.show()  
Montecarlo()    


   
```
