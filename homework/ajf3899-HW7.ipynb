```python
import numpy as np
import matplotlib.pyplot as plt

# Function representing dy/dx = -2y*cos(x)
def f(x, y):
    return -2 * y* np.cos(x)
x_start, x_end = 0, 10*np.pi

#Empty lists to store error and h values
average_errors = []
local_errors = []
h_stored = []
lam = lambda x: -2*np.cos(x)
plt.figure(1)
# Defining parameters
for h in np.arange(0.1, 1, 0.05):
    num_steps = int((x_end - x_start) / h)
    accumulated_error = 0
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    running_error = np.zeros(num_steps + 1)                     
    y_values[0] = 1 # y(0) = 1
    #Checking stability condition
    stable = True
    for x in x_values:
        if abs(1-h*lam(x) >= 1):
            stable = False
            break
    # Implement Euler's method
    for i in range(num_steps):       
        y_values[i + 1] = y_values[i] + h *f(x_values[i], y_values[i])
        current_exact = np.exp(-2 * np.sin(x_values[i + 1]))
        current_error = abs(current_exact - y_values[i + 1])
        accumulated_error += current_error 
        running_error[i + 1] = accumulated_error
    #Finding the local error at x= 15 for each run    
    index_15 = np.argmin(np.abs(x_values - 15))  
    y_approx_15 = y_values[index_15]
    y_exact_15 = np.exp(-2 * np.sin(x_values[index_15]))
    local_error = abs(y_exact_15 - y_approx_15)
    
    local_errors.append(local_error)
    average_error = accumulated_error/num_steps                           
    h_stored.append(h)
    average_errors.append(average_error)
                               
    plt.plot(x_values, y_values)
    print(f"Euler Approximation when h = {h}","------>", y_values[-1])
    print("The current error is:","------>", running_error[-1] )
    print(f"The Accumulated error is: {accumulated_error}", '\n')
        

# Plot the results                              
#Euler approximation
print("---Euler Approximation---", '\n')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2ycos(x) using Euler's Method")

#Average error vs h
plt.figure(2)
plt.plot(h_stored, average_errors, color = 'k', marker = '*')
plt.xlabel('h')
plt.ylabel('Average Error')
plt.title("Average Error between runs of Euler's method")
plt.show() 

#Local error vs h
plt.figure(3)
plt.plot(h_stored, local_errors, color = 'y', marker = '^')
plt.xlabel('h')
plt.ylabel('Local Error')
plt.title("Local Error at x =15 between runs of Euler's method")
plt.show() 
                            
```

```python
#DICUSSION OF RESULTS

print("Since Euler's method assumes that the slope of the function is constant over a small step size, and because we are attempting to solve a sinusoidal differential equation, the first graph will look inherently unstable")

print('\n'"As for a comparison between the local vs average global error, we can clearly notice how local errors behave with periodic oscillations that begin increasing over h because it is assuming that the perious step was correct. Average error, in the other hand, seems to be a parabola-like trend where the average error peaks at the equinox of h values- ultimately spiraling away from the true values as h increases (probably because of the tradeoff between increasing the error per step and decreasing number of steps as h increases)")
```
