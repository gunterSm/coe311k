```python
#Jacobi method
import time
import numpy as np

A = np.array([
 [0.582745, 0.48, 0.10 , 0. , 0.  ], 
 [0.48, 1.044129, 0.46, 0.10, 0.],
 [0.10, 0.46, 1.10431 , 0.44, 0.10 ],
 [0., 0.10, 0.44, 0.963889, 0.42 ],
 [0., 0., 0.10 , 0.42 , 0.522565]], dtype=float)

diag = abs(np.diag(A))
row_sums = np.sum(abs(A), axis=1)
Is_dominant = True 

for i in range(len(A)):
    if ((diag[i] <= row_sums[i]-diag[i])):
        Is_dominant = False
if Is_dominant:
    print(f" This matrix is diagonally dominant: \n {A}")   
else:
    print("This matrix is NOT diagonally dominant")
```
