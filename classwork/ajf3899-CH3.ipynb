```python
import random
import matplotlib.pyplot as plt
import numpy as np

def Montecarlo(Samples):
        dots = 0      
        for i in range(int(Samples)):
            x = random.uniform(-1, 1)
            y = random.uniform(-1, 1)
            if((x*x + y*y) < 1):
                dots += 1
            
        pi_estimation= 4.0 *(dots/ i)
        return(pi_estimation)
        
```

```python
RelativeError = []
Error = []
Iterations = []

for i in range(1,101):
    samples = 100 * i
    calculated_pi = Montecarlo(samples)
    true_pi = np.pi
    error = abs(np.pi - calculated_pi) 
    Error.append(error)
    relative_error = abs((true_pi-calculated_pi)/true_pi)
    RelativeError.append(relative_error)
    Iterations.append(samples)
    
plt.subplot(2, 1, 1)
plt.plot(Iterations, Error, 'k')
plt.ylabel("True Error")
plt.title("Monte Carlo Ï€ Error")
plt.subplot(2, 1, 2)
plt.plot(Iterations, RelativeError, 'g')
plt.xlabel("Number of Samples")
plt.ylabel("Relative Error")
plt.show()
```
