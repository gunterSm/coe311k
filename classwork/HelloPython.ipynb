```python
import random 

def approximateMontecarlo(numberSamples):
    counter = 0
    
    for i in range(0,numberSamples):
        x =random.uniform(-1,1)
        y =random.uniform(-1,1)

    if (x * x + y * y < 1):
        counter = counter +1
        
    return 4.0 * (counter /numberSamples);
```

```python
py = approximateMontecarlo(350000)
print(py)
```
