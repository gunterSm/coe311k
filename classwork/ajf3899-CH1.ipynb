```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0, 2 * np.pi, 100)

plt.subplot(2,1,1)
plt.plot(x, np.sin(x), 'r')
plt.show()
plt.subplot(2,1,1)
plt.plot(x, np.cos(x), 'g')
plt.show()
plt.subplot(2,1,1)
plt.plot(x, np.arcsin(x), 'b')
plt.show()
plt.subplot(2,1,1)
plt.plot(x, np.arccos(x), 'm')
plt.show()
```
