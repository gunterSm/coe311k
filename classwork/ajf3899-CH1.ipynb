```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0, 2 * np.pi, 100)
s = np.sin(x)
y = np.cos(x)
m = np.tan(x)

#Regular sin(x) graph contained within 0 to 2pi

plt.plot(x, s, 'g')

#Histogram of cos(x) 

plt.hist(y, 10, color = 'b')
#Scatterplot of the cos(x) values contained within the x parameters 
plt.scatter(x,y, color = 'k')

#Stem is apparently a type of graph

plt.stem(x, m, 'y')
plt.show()
```
