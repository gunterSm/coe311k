```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x, y):
    return -1000*y +3000-2000*math.exp(-x)
x_start, x_end = 0, 30
def exact(x):
    return 3-0.998*np.exp(-1000*x)-2.002*np.exp(-x) #Apparently I can't use math.exp with arrays, so I used np

# Define parameters
h_val = [.1, .01, .001]

for h in (h_val):
    num_steps = int((x_end - x_start) / h)
    accumulated_error = 0
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    running_error = np.zeros(num_steps + 1)                     
    y_values[0] = 0 # y(0) = 0

    # Implement Euler's method
    for i in range(num_steps):                      
        y_values[i + 1] = y_values[i] + h *f(x_values[i], y_values[i])
    plt.plot(x_values, y_values, label = f'Eulers forward with h ={h}')

# Plot the results
t_exact = np.linspace(x_start, x_end, 5000)
y_exact = exact(t_exact)
plt.plot(t_exact, y_exact, '--', label='Exact solution', color = 'k')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Eulers forward vs. Analytical Solution")
plt.legend()
plt.show()
```
