```python
import numpy as np
import matplotlib.pyplot as plt

#I believe we are plotting this between an interval of 0 to 2pi
f = lambda x: np.sin(x)
a, b = 0, 2*np.pi
#I wanted to use 20 sample points because 50 would make the graphs look crowded and dont show the differences between graphs
N=  20
#Uniform Grid of x going to 0 to 2pi with 20 evenly spaced points 
x = np.linspace(a, b, N)
y= f(x)

#Step size
h = (b-a)/(N-1)

#Central Difference Approximation

df_dx_central = (f(x+h)- f(x-h))/ (2*h)

#Forward Difference Approximation

df_dx_forward = (f(x+h)-f(x))/ h

#Backward Difference Approximation

df_dx_backward = (f(x)- f(x-h))/ h

#Plotting
plt.grid(color='gray', linestyle='--', linewidth=1) #Got this from google
plt.plot(x, np.cos(x), '--', color = 'y', label='Exact Derivative')
plt.plot(x, df_dx_central, '*--',color = 'g', label='Central Difference Approximation')
plt.plot(x, df_dx_forward, 'o--', color = "b", label='Forward Difference Approximation')
plt.plot(x, df_dx_backward, '^--', color = "c", label='Backward Difference Approximation')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Approximating Derivative of f(x) = sin(x) using CentralDifference")
plt.legend()
plt.show()
```
