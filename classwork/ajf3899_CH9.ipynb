```python
#Challenge Part 1
import numpy as np
import matplotlib.pyplot as plt

# Data points

X = np.array([1, 2, 3, 4, 5])

Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")


plt.scatter(X,Y, marker = 'o', color = 'g')
plt.plot(X, m*X+c, linestyle= 'solid')
plt.title("Best Fitting Line")
plt.show()
```

```python
#Challenge Part 3
import numpy as np
import matplotlib.pyplot as plt

#My equation is y = 3.5x + 8

s = 3.5
t = 8

X = np.arange(0, 1, 0.01)

Y = s*X + t

#The noise
noise = 0.5*(np.random.rand(len(X))- 0.5)

# Stack the design matrix with a column of ones (for the intercept)
y = Y - noise

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")


plt.scatter(X,y, marker = '^', color = 'c')
plt.plot(X, m*X+c, linestyle= 'solid', color = 'k')
plt.title("Best Fitting Line")
plt.show()
```
