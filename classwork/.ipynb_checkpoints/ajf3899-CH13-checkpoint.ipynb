```python
import numpy as np
import random 
import time

#Pair 1
a = np.random.randn(3, 3)
b = np.random.randn(3, 3)
#Pair 2
c = np.random.randn(6, 6)
d = np.random.randn(6, 6)
#Pair 3
e = np.random.randn(12, 12)
f = np.random.randn(12, 12)
#Pair 4
g = np.random.randn(24, 24)
h = np.random.randn(24, 24)
#Pair 5
i = np.random.randn(48, 48)
j = np.random.randn(48, 48)
#Pair 6
k = np.random.randn(96, 96)
l = np.random.randn(96, 96)
#Pair 7
m = np.random.randn(192, 192)
n = np.random.randn(192, 192)
#Pair 8
o = np.random.randn(384, 384)
p = np.random.randn(384, 384)


def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range (n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

#Multiplying pairs

#a with b
start_time = time.perf_counter()  
ans1 = naive_matrix_multiplication(a,b)
end_time = time.perf_counter()

elapsed_time1 = end_time - start_time

print("This is the elapsed time for pair 1 multiplication: ", '\n', elapsed_time1)

#c with d 
start_time = time.perf_counter() 
ans2 = naive_matrix_multiplication(c,d)
end_time = time.perf_counter()

elapsed_time2 = end_time - start_time

print("This is the elapsed time for pair 2 multiplication: ", '\n', elapsed_time2)

#e with f 
start_time = time.perf_counter()
ans3 = naive_matrix_multiplication(e,f)
end_time = time.perf_counter()

elapsed_time3 = end_time - start_time

print("This is the elapsed time for pair 3 multiplication: ", '\n', elapsed_time3)

#g with h
start_time = time.perf_counter()
ans4 = naive_matrix_multiplication(g,h)
end_time = time.perf_counter()


elapsed_time4 = end_time - start_time

print("This is the elapsed time for pair 4 multiplication: ", '\n', elapsed_time4)

#i with j
start_time = time.perf_counter()
ans5 = naive_matrix_multiplication(i,j)
end_time = time.perf_counter()

elapsed_time5 = end_time - start_time

print("This is the elapsed time for pair 5 multiplication: ", '\n', elapsed_time5)

#k with l
start_time = time.perf_counter()
ans6 = naive_matrix_multiplication(k,l)
end_time = time.perf_counter()

elapsed_time6 = end_time - start_time

print("This is the elapsed time for pair 6 multiplication: ", '\n', elapsed_time6)

#m with n
start_time = time.perf_counter()
ans7 = naive_matrix_multiplication(m,n)
end_time = time.perf_counter()

elapsed_time7 = end_time - start_time

print("This is the elapsed time for pair 7 multiplication: ", '\n', elapsed_time7)

#o with p
start_time = time.perf_counter()
ans8 = naive_matrix_multiplication(o,p)
end_time = time.perf_counter()

elapsed_time8 = end_time - start_time

print("This is the elapsed time for pair 8 multiplication: ", '\n', elapsed_time8)
```

```python
#Np multiplication version 

#First one
start_time = time.perf_counter()
m1 =  a@b
end_time = time.perf_counter()

elapsed_time_1 = end_time - start_time

print("This is the elapsed time for pair 1 multiplication: ", '\n', elapsed_time_1)

#Second one 
start_time = time.perf_counter()
m2 =  c@d
end_time = time.perf_counter()

elapsed_time_2 = end_time - start_time

print("This is the elapsed time for pair 2 multiplication: ", '\n', elapsed_time_2)

#Third one 
start_time = time.perf_counter()
m3 =  e@f
end_time = time.perf_counter()

elapsed_time_3 = end_time - start_time

print("This is the elapsed time for pair 3 multiplication: ", '\n', elapsed_time_3)

#Fourth one 
start_time = time.perf_counter()
m4 =  g@h
end_time = time.perf_counter()

elapsed_time_4 = end_time - start_time

print("This is the elapsed time for pair 4 multiplication: ", '\n', elapsed_time_4)

#Fifth one 
start_time = time.perf_counter()
m5 =  i@j
end_time = time.perf_counter()

elapsed_time_5 = end_time - start_time

print("This is the elapsed time for pair 5 multiplication: ", '\n', elapsed_time_5)

#Sixth one 
start_time = time.perf_counter()
m6 =  k@l
end_time = time.perf_counter()

elapsed_time_6 = end_time - start_time

print("This is the elapsed time for pair 6 multiplication: ", '\n', elapsed_time_6)

#Seventh one 
start_time = time.perf_counter()
m7 =  m@n
end_time = time.perf_counter()

elapsed_time_7 = end_time - start_time

print("This is the elapsed time for pair 7 multiplication: ", '\n', elapsed_time_7)

#Eigth one 
start_time = time.perf_counter()
m8 =  o@p
end_time = time.perf_counter()

elapsed_time_8 = end_time - start_time

print("This is the elapsed time for pair 8 multiplication: ", '\n', elapsed_time_8)
```

```python
#Plotting
import matplotlib.pyplot as plt 

Run_times_class_algorithm = [elapsed_time1, elapsed_time2, elapsed_time3, elapsed_time4, elapsed_time5, elapsed_time6, elapsed_time7, elapsed_time8]
Run_times_pi_algorithm = [elapsed_time_1, elapsed_time_2, elapsed_time_3, elapsed_time_4, elapsed_time_5, elapsed_time_6, elapsed_time_7, elapsed_time_8]

matrix_sizes = [3, 6, 12, 24, 48, 96, 192, 384]
plt.plot(matrix_sizes, Run_times_class_algorithm, marker = '*', color = 'y', label = "Naive")
plt.plot(matrix_sizes, Run_times_pi_algorithm, marker = 'o', color = 'b', label = "Optimized")
plt.xlabel("Matrix size")
plt.ylabel("Run time (in seconds)")
plt.legend()

#The times for the naive version are much greater than the optimized pi matrix multilpication library 
#The optimized version(pi library) is better, it has a smaller Big 0 notation and it is less computationally expensive than the naive matrix multiplication  
```
