```python
#Solving a Second ODE using Euler's method, first part 
import numpy as np
import matplotlib.pyplot as plt
# Define the parameters
k = 5.0 # Constant in the equation (e.g., springconstant), which I modified to be 5
h_val = [.01, .02, .03, .05, .1, .2, .3, .5] # Step size, idk if this was asked tho...
x_start = 0 # Start of the interval
x_end = 10 # End of the interval
y_0 = 1.0 # Initial condition for y
v_0 = 0.0 # Initial condition for dy/dx
# Create arrays to store x, y, and v values
for h in h_val:
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
# St initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

# Euler's Method loop
    for i in range(1, len(x_values)):
    # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
# Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
    plt.plot(x_values, y_values, label = f"h value is = {h}")
# Plot the results
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.grid(True)
plt.legend()
plt.show()

print('In this case, the graph seems to become unstable when h is around 0.2 and k =5- as observed from the big oscillations that h= 0.2, 0.3, and 0.5 present. As x goes on, it seems that starting from h = 0.2 and all our following h-values are becomimg more unstable.')
print('\n''I also noticed that values of k less than 5 make all the graphs to present large fluctuations- which is weird because values of k larger than 5 doesnt seem to fix the oscillations as increasing the value from 1 to 5 did for the values of h less than 0.2.  ')
```

```python
#Solving a Second ODE using Euler's method, second part 
import numpy as np
import matplotlib.pyplot as plt
# Define the parameters
k = 5.0 # Constant in the equation (e.g., springconstant), which I modified to be 5
h = 0.01 # Step size
x_start = 0 # Start of the interval
x_end = 10 # End of the interval
y_0 = 1.0 # Initial condition for y
v_0 = 0.0 # Initial condition for dy/dx
# Create arrays to store x, y, and v values
x_values = np.arange(x_start, x_end + h, h)
y_values = np.zeros(len(x_values))
v_values = np.zeros(len(x_values))
# Set initial conditions
y_values[0] = y_0
v_values[0] = v_0

# Euler's Method loop
for i in range(1, len(x_values)):
# Current values
    x_n = x_values[i - 1]
    y_n = y_values[i - 1]
    v_n = v_values[i - 1]
# Update equations
    y_values[i] = y_n + h * v_n
    v_values[i] = v_n + h * (-k * y_n)
# Plot the results
plt.figure(figsize=(18, 5))
plt.scatter(x_values, y_values, marker = '.', color = 'k')
plt.plot(x_values, y_values, label='Position y(x)')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()

print('I dont believe that Eulers method cannot solve this problem with bigger h because as h increases the graph seems to become more unstable. It should be solved with smaller step sizes, but as we argued in class, that would be computationally expensive')
```
