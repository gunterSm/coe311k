```python
'''
EXAMPLE IN CLASS
using a for loop, find the triples that satisfies
a*a + b*b = c*c
where
0 < a < 100
0 < b < 100'''

import math

for a in range(1,100):
    for b in range(1,100):
        c= math.sqrt(a**2+b**2)
        if(int(c) == c):
            print("Here are your triplets:")
            print(a,b, int(c))
            
            
                


   
```

```python
x = [i for i in range(0,101)]
y_1 = []
y_2 = []
for i in x:
    y_1.append(i**2)
for i in x:
    y_2.append(i**3)

with open("output.txt", "w") as f:
    for i in range(len(x)):
        f.write(f"{x[i]}, {y_1[i]}, {y_2[i]}\n")
```
