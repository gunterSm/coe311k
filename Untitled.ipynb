 ## SECTION 1: INTRODUCTION AND SYSTEM SELECTION

# The system we are using is a mechanical system with high damping, specifically,
# an overdamped spring-mass system where the mass is negligible compared to the
# damping and spring forces. A strut for airplane landing gear falls under this
# category, as it tries to minimize large oscillations as quickly as possible.
#
# **The governing ODE**:
#
# dx/dt = -5000x + 0.002sin(t),  with initial condition x(0) = 1 m
#
# This comes from the standard form dx/dt = -(k/c)x + (1/c)F(t), where k/c = 5000 s^-1
# is the ratio of the spring constant to the damping coefficient, and F(t) = 10sin(t) N
# is a slowly varying external force. The 1/c factor scales the force term down to 0.002.
#
# **Physical meaning of variables and parameters**:
#
# x(t) is the displacement of the system in meters, starting at x(0) = 1 m, meaning
# the strut begins compressed by one meter at the moment of contact. The -5000x term
# drives the system rapidly back toward equilibrium; this is the dominant fast dynamics.
# The 0.002sin(t) term represents the slow external forcing. It oscillates on the scale
# of seconds, while the damping acts on the scale of milliseconds. That separation of
# time scales is exactly what makes this equation stiff.
#
# **Why this system is important in practice**:
#
# Getting the damping right matters a lot in practice. If the system is overdamped,
# the strut compresses too slowly, and the full landing load gets transferred to the
# airframe as a hard jolt - at best uncomfortable, at worst structurally damaging.
# If it is underdamped, the strut bounces and the aircraft oscillates on the runway,
# which is a handling and safety problem. The stiffness of this ODE is not just a
# numerical curiosity; it reflects the real physical challenge of a system that snaps
# to equilibrium in milliseconds while being driven by forces that vary over seconds.

# ## SECTION 2: DEMONSTRATION OF STIFFNESS
#
# By applying Euler's Forward to dx/dt = -5000x + 0.002sin(t), we can see immediately
# that the solution blows up for any step size that would be considered "reasonable"
# for a 5-second simulation. The method requires an impractically small step size just
# to stay stable, not because we need that precision, but purely to keep the numerics
# from diverging. That is the defining characteristic of a stiff equation.
#
# **Stiffness ratio**:
#
# For this problem, stiffness comes from the mismatch between the two time scales present.
# The fast dynamics are governed by the decay rate λ = 5000 s^-1, which means the
# homogeneous part of the solution decays with a time constant of τ_fast = 1/5000 = 0.0002 s.
# The slow dynamics come from the forcing term sin(t), which has a frequency of 1 rad/s
# and varies on a time scale of τ_slow ~ 1 s. The stiffness ratio is therefore:
#
#     τ_slow / τ_fast = 1 / 0.0002 = 5000
#
# A ratio of 5000 is well into stiff territory where explicit methods struggle.
# Note: the scalar stiffness ratio λmax/λmin = 1 is not meaningful here because there
# is only one eigenvalue. The relevant comparison is always between the fast decay
# and the slow forcing, not between eigenvalues of the same term.
#
# **Stability condition for Euler's Forward**:
#
# For Euler's Forward to remain stable on this problem, we need |1 + hλ| < 1,
# where λ = -5000. Substituting and solving for h:
#
#     |1 - 5000h| < 1
#     -1 < 1 - 5000h < 1
#     0 < h < 0.0004
#
# So the maximum stable step size is h = 0.0004 s. To simulate 5 seconds at this
# step size requires 5 / 0.0004 = 12,500 steps just to maintain stability, not
# accuracy. That is computationally impractical for a problem where h = 0.1 would
# be perfectly sufficient for the slow dynamics. This is exactly why implicit methods
# are necessary here.
