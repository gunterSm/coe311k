## SECTION 1: INTRODUCTION AND SYSTEM SELECTION


The system we are using is a mechanical system for high damping. A strut for airplane landing gear falls under this category, as it tries to minimize the large oscillations quickly.

**The governing ODEs**:

the ODE is (dx/dt) = -5000x + 0.002sin(t)with the initial conditions that X0 = 1m

**Physical meaning of variables and parameters**: 

This means that the position is initially one meter up, and that the damping ratio is -5000. This is high damping, so the system should stop oscillating pretty quickly---this is consistent with the motion of an aircraft landing. The second term is in reference to the force that the damping system is counteracting. The forces, if compared with the reaction of the damping, occurs slowly, oscillating on the magnitude of seconds, not miliseconds.

**Why this system is important in practice**: 

This system is important because there is a balance that is necessary to hit between overdampened and underdampened. If the system is overdampened, then the plane will, at worst, be severely damaged from the force of landing, at best, experience an uncomfortable jolt for the passengers/precious cargo. However, underdampening the system results in, at worst, a crash because the body of the plane will hit the ground, or at best, an uncomfortable swaying up and down that nauseates the passengers.

##  SECTION 2: DEMONSTRATION OF STIFFNESS

By applying Euler's Forward to the linear equation dx/dt = -5000x + 0.002sin(t), we notice that the graphed solutions are not accurately modeling the function throughout the time interval- which means we need much tinier step sizes for the graph to become stable. As a result, using explicit methods like Euler's Forward is not a practical tool in this scenario.

Due to this behavior, we might need to solve for the stiffness ratio-which tells you how different the speeds of the system are- to find if we were given a stiff equation:

From our dx/dt equation, $\lambda=-5000$, and since it is the only λ(lamda), it would be both our maximum and minimum value for lamda.

The equation for stiffness ratio is |λmax| / |λmin|; plug in λmax and λmin, |-5000|/|-5000| = 1. 

A value of 1 doesn't tell us much because there is no fast vs. slow comparison but only one behavior. 

**So how do we know we are dealing with a stiff equation?**

We can implement Euler's stability condition, where ∣1+hλ∣<1. Substituting for λ and solving for h:

−1< 1 − 5000h <1

−2< − 5000 h <0

0<  h  < 0.0004

This value means that for a given time interval, 5 in this case, we would need 12500 steps to accurately depict the behavior during the time interval for this function- which is unviable. Contrary to explicit methods, implicit methods like Euler's backwards ensure numerical stability while allowing much larger step sizes when solving stiff equations. 

```python
import numpy as np
import matplotlib.pyplot as plt

h_vals = [0.1, 0.01, 0.001]
ratio = 5000

t_start, t_end = 0, 5

def f(t):
    return 0.002*np.sin(t)

def dx(t, x):
    return -ratio*x + f(t)

for h in h_vals:
    num_steps = int((t_end - t_start) / h)

    t_values = np.linspace(t_start, t_end, num_steps + 1)
    x_values = np.zeros(num_steps + 1)

    x_values[0] = 1

    for i in range(num_steps):
        x_values[i + 1] = x_values[i] + h * dx(t_values[i], x_values[i])

    print(f"\nh = {h}")
    print(f"First 3 x-values: {x_values[:3]}")
    print(f"Final x-value: {x_values[-1]}")
    print(f"Max |x| value: {np.max(np.abs(x_values))}")

    plt.plot(t_values, x_values, label=f"h = {h}")

plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Euler Forward for dx/dt = -5000x + 0.002sin(t)")
plt.legend()
plt.grid(True)
plt.show()
```

## SECTION 3: MATHEMATICAL SETUP

*Euler's backward* says that: yn+1 =yn + hf(tn+1 ,yn+1)

Move everything to one side, 0 = yn+1 - yn - h(f(tn+1 ,yn+1)), an equation which we would name g(yn+1). 

For our equation f(t,y)=−5000 y+ 0.002 sin(t). Plugging in our function f into Euler's Backward yields:

g(yn+1) = yn+1 - yn -h(-5000yn+1 + 0.002 sin(tn+1))

g(yn+1) = yn+1 - yn +5000h yn+1 - 0.002hsin(tn+1)

The result is a function that measures how far is our guess from satisfying the numerical method. 

**To find the residuals:**

We need to take the derivative of our equation g(y) with respect to y. Doing so would yield:

g'(yn+1)= 1 - 0 + 5000h - 0

g'(yn+1)= 1 + 5000h

The result is a function that measures how the residual changes when the guess for the next value also changes.

We need the *Newton Raphson* update formula in order to update our interations, which is: 

yn+1 ^(k+1) =  yn+1^(k)  - g( yn+1^(k) ) / g'( yn+1^(k) ), plugging in our found equations yields:

yn+1 ^(k+1) =  yn+1^(k)  - (yn+1^(k) - yn +5000h yn+1^(k) - 0.002h sin (t n+1)) / 1 + 5000 h





## SECTION 4: IMPLEMENTATION


```python
import numpy as np
import matplotlib.pyplot as plt 

# Parameters
x0 = 1 # Initial condition
t0 = 0
tf = 5
h = 0.1

#the function
def f(t, x):
  return (-5000*x) + (0.002* np.sin(t))

#the ODE
def dfdx(t,x):
  return -5000

#Newton-Raphson
def newton_raphson(g, dg, x0, tol=1e-6, max_iter=20):
  x = x0
  for i in range(max_iter):
    x_new = x - g(x) / dg(x)
    if abs(x_new - x) < tol:
      return x_new, i + 1
    x = x_new
  return None, max_iter

# Backward Euler with Newton-Raphson
def backward_euler_newton(f, dfdx, y0, t0, tf, h):
    t_values = np.arange(t0, tf + h, h)
    y_values = np.zeros_like(t_values)
    y_values[0] = y0

    iteration_counts = []
    for i in range(1, len(t_values)):
        t_next = t_values[i]
        y_prev = y_values[i - 1]

        # Define g and dg
        g = lambda y: y - y_prev - h * f(t_next, y)
        dg = lambda y: 1 - h * dfdx(t_next, y) #chose to keep the original code despite linearity

        # Use defined Newton method
        y_guess = y_prev + h * f(t_values[i-1], y_prev)
        y_next, iters = newton_raphson(g, dg, y_guess)

        iteration_counts.append(iters)
        y_values[i] = y_next
    return t_values, y_values, iteration_counts

# Solve the ODE
h_values = [0.1,0.01,0.001]

for h in h_values:
    t_vals, y_vals, iters = backward_euler_newton(f, dfdx, x0, t0, tf, h)
    print(f"h = {h}, avg Newton iterations = {np.mean(iters)}")
 
```

**B) COMPARISON PLOTS SHOWING SOLUTION BEHAVIOR**






```python
#For this part, we attempted to code a graph plotting both method solutions
#however, this was not possible because the data blows up for Euler's backwards and doesn't provide any useful visual comparison.

#We are citing ChatGPT on this part with the prompt: How do you create two separate graphs that are iterating over different parameters.
#Ultimately, we incorporated the AI advice into our set of equations. 
import numpy as np
import matplotlib.pyplot as plt

h_vals = [0.1, 0.01, 0.001]
t0, tf = 0, 5
x0 = 1

def force(t):
    return 0.002*np.sin(t)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

for h in h_vals:
    t = np.arange(t0, tf + h, h)

    xF = np.zeros(len(t))
    xB = np.zeros(len(t))
    xF[0] = x0
    xB[0] = x0

    for i in range(len(t)-1):
        # Forward Euler
        xF[i+1] = xF[i] + h*(-5000*xF[i] + force(t[i]))

        # Backward Euler
        xB[i+1] = (xB[i] + h*force(t[i+1])) / (1 + 5000*h)

    axs[0].plot(t, xF, label=f"h={h}")
    axs[1].plot(t, xB, label=f"h={h}")

axs[0].set_title("Forward Euler")
axs[0].set_xlabel("Time (s)")
axs[0].set_ylabel("x(t)")
axs[0].grid(True)
axs[0].legend()

axs[1].set_title("Backward Euler")
axs[1].set_xlabel("Time (s)")
axs[1].set_ylabel("x(t)")
axs[1].grid(True)
axs[1].legend()

plt.suptitle("Euler Method Comparison for Stiff ODE")
plt.tight_layout()
plt.show()
```

**c) NEWTON-RAPHSON CONVERGENCE ANALYSIS**:

The convergence behavior is consistent with what one should expect from a linear system. Because the highest term is linear, g is linear, and then dg, the derivative is a constant. With a constant dg, this method is always able to converge in 1 step, because it is easily able to find the exact root of the function in 1 step.


```python
for h in h_vals:
    t_vals, y_vals, iters = backward_euler_newton(f, dfdx, x0, t0, tf, h)
    plt.plot(t_vals[1:], iters, label=f"h = {h}")

plt.xlabel("time(seconds)")
plt.ylabel("Iterations")
plt.title("Newton Raphson Iterations v Time")
plt.legend()
plt.grid(True)
plt.show()
```

## SECTION 6: PERFORMANCE ANALYSIS

**Computational cost analysis**: 

A stiff method, while it does involve more complicated code through the Newton-Raphson method, is actually a more efficient method than the forward-eulers method. Forward euler's is a simple method, however, it is limited to what it can be used for. For stiff situations, then forward eulers demands a very small step size; then dragging the system down. Moreover, at an insufficiently small step size, the system completely blows up. For this reason, the backwards euler using Newton-Raphson has a lower computational cost, because it allows for a larger step size.

**Table comparing Euler's Forward vs. Backward**:


```python
import matplotlib.pyplot as plt
import pandas as pd

table_solutions_forward = pd.DataFrame({
    "Step Size (h) ": ['0.1', '0.01', '0.001'],
    "Solution": ['8.03 e134', 'nan', 'nan']
})

table_solutions_forward

```

```python
print("SOLUTIONS FOR EULER'S BACKWARD FOR VARYING STEP SIZES")

table_solutions = pd.DataFrame({
    "Step Size (h) ": ['0.1', '0.01', '0.001'],
    "Solution": ['1.08', '1.01', '1.0018']
})

table_solutions
```

Both methods solutions can be seen here. From the first table it is clear that at the largest step size, the solution blows up to a gigantic number that is completely wrong because euler's forward needs more precision to operate. However, from the same table, it is clear that with a smaller step size, we receive a run time error, where the code becomes too sluggish to work. In contrast, the euler's backward with NR operates at almost the same quality for step sizes 0.1, 0.01, and 0.001. This shows that we can get just about the same result at 0.1 as 0.001. This is a huge difference, and makes our implicit method significantly more efficient.

**Discussion of accuracy-efficiency tradeoff**: 

While Euler's backward excels in both efficiency and accuracy for a stiff scenario, not all step sizes are the same. While I can get virtually the same solution from any step size, as step size increases, the accuracy decreases. It may be a minor change, however, if accuracy matters then a smaller step size will still be preferable to a larger step size.

**Practical recommendations**: 

Of course, each of these choices is context-dependent. For some contexts, accuracy will matter more. For the predictions of the motion of an airplane strut, accuracy is important, because it carries a weight of human life and the damage of very expensive machinery. A step size of 0.001 will be sufficient to analyze any design changes.

## SECTION 7: CONCLUSIONS



**Summary of key findings**:

What we have seen over the course of analyzing the motion of our dampened strut is that: 1) this is a stiff system, 2) Euler's forward cannot properly analyze this because of the competing motion of the two terms, and 3) Euler's backward with the use of Newton-Raphson is capable of analyzing this stiff system. First, it is clear that our dampened system is a stiff system by simplying considering the structure of the differential equation:  (dx/dt) = -5000x + 0.002sin(t). We have the first term of -5000x, this is a constantly increasing expression. The second term; 0.002sin(t), is very fast oscillating motion. This combined set up is generally considered a stiff equation. Secondly, based on the solutions---or lack thereof---from Euler's forward method, it is clear that this is a failure of the method. It is necessitating step sizes that are too small to function. This is computationally ineffective and very impractical. Third, by the contrasting solutions provided by the Euler's backwards with Newton-Raphson method, it is obvious that analyzing stiff equations is a strong suit of this method. It provides a quick and accurate solution to this system.

**When to use implicit vs. explicit methods**

Implicit methods are used when the situation you are concerned with is considered "stiff". This is to say, it is comprised of at least two functions, where one is changing very gradually, and the other is changing very rapidly. Explicit methods are used anywhere else you can. For many applications, an explicit method will be sufficient. It especially thrives in situations with gradual and smooth movements, where a small step size is possible.


**Challenges encountered and lessons learned**

We encountered some challenges when trying to graph our stiff equation using Euler's Forward, where we had to go find the h step size we could use because even really small values were too large for this method- where the largest h value we could use was 0.0004. So we learned that explicit methods are useful for solving short-duration dynamic events, such as crashes, explosions, or blast impacts; but for our strut dampening analysis, we would need a more efficient method that could handle larger step sizes. 

This is where implicit methods are extremely useful because they can handle big step sizes while being efficient. They are unconditionally stable, but at the cost of being computationally expensive due to the matrix inversion for each step- which is something else that we learned: that there is always tradeoffs and specific scenarios where each method could be useful, and if not, there must be a better, less computationally expensive method out there that could be applied into our scenario.
