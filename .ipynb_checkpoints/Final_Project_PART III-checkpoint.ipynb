## SECTION 1: REVIEW OF PART 2

In Part 2, we analyzed a *stiff system* modeling a highly damped mechanical system, such as an aircraft landing gear strut. The governing ODE is:

dx/dt = -5000x + 0.002sin(t)

- with initial condition x(0) = 1.

The large negative coefficient (-5000) represents very strong damping, which causes the system to decay extremely quick. In contrast, the forcing term 0.002sin(t) varies slowly over time. This difference in time scales is what makes the system stiff.

We showed that Euler’s Forward method fails unless extremely small step sizes are used, due to the stability condition:

0 < h < 0.0004

This makes explicit methods impractical. To address this problem, we implemented Euler’s Backward method with Newton-Raphson, which remains stable for much larger step sizes.

## SECTION 2: NEWTON-RAPHSON DAMPING IMPLEMENTATION


Basic Newton-Raphson works well when the initial guess is close to the solution. However, it can fail when:

- The step size h is large  
- The initial guess is poor  
- The update overshoots the solution  

To fix this, we introduce damping using a factor α.

Instead of taking the full Newton step:

y_new = y - g(y) / g'(y)

we modify the update to:

y_new = y + α · Δ

where:

Δ = -g(y) / g'(y)

and α ∈ (0, 1].

The damping factor α is reduced (typically by halving) until the residual decreases, ensuring that each iteration moves toward a more stable solution rather than diverging- where damping is included as a robustness feature for larger step sizes or more nonlinear implicit systems. 

As shown in the results, both methods converged in the same number of iterations, 1,  to the same computed value, 0.0001 were damping is included as a safeguard for nonlinear problems where standard Newton-Raphson may fail (although it wasn't used because we are using a linear equation)

```python
#Implementation of damped NR solver
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ODE
# -5000 term makes it decay fast (stiff)
def f(t, x):
    return -5000*x + 0.002*np.sin(t)

def dfdy(t, x):
    return -5000

# g and g'
def g_func(y, y_n, t_next, h):
    return y - y_n - h * f(t_next, y)

def g_prime_func(h):
    return 1 + 5000*h

#Solver 
def damped_newton_raphson(y_n, t_next, h, tol=1e-8, max_iter=20, alpha_min=1e-4):
    y_guess = y_n
    damping_used = False
    total_backtracks = 0

    for k in range(max_iter):
        g = g_func(y_guess, y_n, t_next, h)

        if abs(g) < tol:
            return y_guess, k, True, damping_used, total_backtracks

        g_prime = g_prime_func(h)
        delta = -g / g_prime

        alpha = 1.0
        g_current = abs(g)

        # damping loop: if step makes things worse, shrink it
        for _ in range(10):
            y_new = y_guess + alpha * delta
            g_new = abs(g_func(y_new, y_n, t_next, h))

            if g_new < g_current:
                break

            alpha /= 2.0
            damping_used = True
            total_backtracks += 1

            if alpha < alpha_min:
                return y_guess, k, False, damping_used, total_backtracks

        y_guess = y_new

    # if we never converge, return failure
    return y_guess, max_iter, False, damping_used, total_backtracks

#---------------------------------------------------------------------------------------------------------------------

#Demonstration where damping helps, using a large step size to test the robustness of damped Newton-Raphson
h_large = 2.0
y0 = 1
t0 = 0
t_next = h_large

print("Testing damped NR with large h......", '\n')

y, iters, success, damp_used, backs = damped_newton_raphson(y0, t0 + h_large, h_large)

print(f"Converged: {success}")
print(f"Iterations: {iters}")
print(f"Damping used: {damp_used}")
print(f"Backtracking steps: {backs}")

#Comparison: basic NR vs. damped NR convergence

def basic_NR(y_n, t_next, h, tol=1e-8, max_iter=20):
    y_guess = y_n
    for i in range(max_iter):
        g = g_func(y_guess, y_n, t_next, h)

        if abs(g) < tol: #Is the current guess close enough to the actual solution? using the tolerance value provided 
            return y_guess, i, True

        g_prime = g_prime_func(h)
        delta = -g / g_prime

        # New guess, y_guess, is equal to the old guess + the correction factor delta
        y_guess = y_guess + delta
    return y_guess, max_iter, False

#Running both methods:
y_basic, it_basic, success_basic = basic_NR(y0, t_next, h_large)
y_damped, it_damped, success_damped, damp_used, backs = damped_newton_raphson(y0, t_next, h_large)

print('-----------------------')
print('\n' 'COMPARISON BETWEEN BASIC NR vs. DAMPED NR CONVERGENCE')

table_convergence = pd.DataFrame({
    "Convergence Method ": ["Basic NR", "Damped NR"],
    "Does it converge?": [success_basic, success_damped],
    "Computed Value of yn+1": [f"{y_basic}", f"{y_damped}"],
    "Number of Iterations": [f"{it_basic}", f"{it_damped}"],
    "Damping Used": ["No", damp_used]
})

table_convergence
```

## SECTION 3: ADAPTIVE STEP SIZE IMPLEMENTATION

For this section, we implemented an adaptive step size method using a step-doubling error estimation technique using the *estimate_error* function. 

This function works by finding a solution at each step in two ways:

- A single step size, h, producing an approximation of yfull
- Two consecutive steps whose size is h/2 that produce a more accurate approximation yhalf.

To find the local error, we can use the difference between these two approximations: 

*error=∣ yhalf −yfull ∣*

For the *step size adjustment*, we used the following logic:

- If the error is smaller than the tolerance, then the step is accepted, and the solution advances to the next iteration.
- If the error is larger than the tolerance, then the step is rejected, and the step size is reduced. 


*Which is this part of the code:* 

if err < tol:
            t.append(t_n + h)
            y.append(y_half)

            h_hist.append(h)
            err_hist.append(err)

            stats['accepted'] += 1
            h = h_new
        else:
            stats['rejected'] += 1
            h = h_new
            
*using the bounds:*

h_new = max(1e-6, min(1.0, h_new))
        h_new = max(0.2 * h, min(5.0 * h, h_new))
        
Ultimately, the goal of implementing a step-doubling error estimation is to provide a reliable error estimate where the difference between the two approximations helps to decide wether to shrink or enlarge the step in order to balance computational cost with accuracy. 

```python
# Estimates error by comparing 1 big step (h) to 2 smaller half steps (h/2)

def estimate_error(y_n, t_n, h):
    # Computes yn+1 using one large step
    y_full, it1 = one_step_NR(y_n, t_n + h, h)
     # If Newton fails, return infinite error np.inf
    if y_full is None:
        return None, None, np.inf, it1
    # Advances solution to midpoint
    y_half_1, it2 = one_step_NR(y_n, t_n + h/2, h/2)
    if y_half_1 is None:
        return None, None, np.inf, it1 + it2
    # Advances solution to the final 
    y_half, it3 = one_step_NR(y_half_1, t_n + h, h/2)
    if y_half is None:
        return None, None, np.inf, it1 + it2 + it3
    # Difference between full step and two half steps
    error = abs(y_half - y_full)
    #f_full is the coarse solution, y_half is the improved solution, the estimated local error, and the iterations is took to measure its computational cost
    
    return y_full, y_half, error, (it1 + it2 + it3)
#-----------------------------------------------------------------------------------------------------------------------------

#Implementation of adaptive step size solver 
def one_step_NR(y_n, t_next, h):
    y, iters, success, _, _ = damped_newton_raphson(y_n, t_next, h)

    if not success:
        return None, iters

    return y, iters

def adaptive_solver(y0, t0, tf, h0, tol):
    t = [t0]
    y = [y0]
    h = h0

    h_hist = []
    err_hist = []

    stats = {
        'accepted': 0,
        'rejected': 0,
        'iters': 0
    }

    max_steps = 100000
    step_count = 0

    while t[-1] < tf and step_count < max_steps:
        step_count += 1

        y_n = y[-1]
        t_n = t[-1]

        y_full, y_half, err, iters = estimate_error(y_n, t_n, h)
        stats['iters'] += iters

        # If Newton failed then shrink step aggressively
        if err == np.inf:
            h = max(0.1 * h, 1e-6)
            stats['rejected'] += 1
            continue

        # Step size re-update
        if err < 1e-14:
            h_new = 2 * h
        else:
            h_new = 0.9 * h * np.sqrt(tol / err)

        # Safety bounds
        h_new = max(1e-6, min(1.0, h_new))
        h_new = max(0.2 * h, min(5.0 * h, h_new))

        # Accept or reject
        if err < tol:
            t.append(t_n + h)
            y.append(y_half)

            h_hist.append(h)
            err_hist.append(err)

            stats['accepted'] += 1
            h = h_new
        else:
            stats['rejected'] += 1
            h = h_new

        # Prevent runaway shrinking, Chat GPT utilzed to understand possible error
        if h < 1e-6:
            print("Warning: minimum step size reached")
            break

    return np.array(t), np.array(y), h_hist, err_hist, stats
#-----------------------------------------------------------------------------------------------------------------------------

#VISUALIZATION
t, y, h_hist, err_hist, stats = adaptive_solver(1, 0, 5, 0.1, 1e-4)

print('\n',"-----------------------------VISUALIZATION OF HOW H ADAPTS DURING THE SOLUTION-----------------------------")
plt.figure(figsize=(15, 15))
plt.subplot(3,1,2)
plt.semilogy(t[:-1], h_hist, 'o-', color ='lightgreen')
plt.title("Step Size h(t)")
plt.xlabel("Time t")
plt.ylabel("Step size (h)")
plt.grid()
```

## SECTION 4: PERFORMANCE ANALYSIS 

```python
import numpy as np
import matplotlib.pyplot as plt

# Solve the ODE from time t0 to tf using a constant step size h
def fixed_solver(y0, t0, tf, h):
    t = np.arange(t0, tf+h, h)
    y = np.zeros_like(t)
    y[0] = y0

    for i in range(len(t)-1):
        y[i+1], _, _, _, _ = damped_newton_raphson(y[i], t[i+1], h)

    return t, y


t_fixed, y_fixed = fixed_solver(1, 0, 5, 0.01)

print("Performance Comparison......", '\n')
print(f"Adaptive steps: {len(t)}")
print(f"Accepted: {stats['accepted']}")
print(f"Rejected: {stats['rejected']}")
print(f"Total NR iterations: {stats['iters']}")

#-------------------------------------------------------------------------------------------------------------------

#PLOTS
t, y, h_hist, err_hist, stats = adaptive_solver(1, 0, 5, 0.1, 1e-4)
plt.figure(figsize=(15, 15))

#Solution with adaptive time points marked
plt.subplot(3,1,1)
plt.plot(t, y, 'o-', color = 'y')
plt.title("Solution")
plt.xlabel("Time t")
plt.ylabel("Solution y(t)")
plt.grid()

#Step size h(t) vs. time
plt.figure(figsize=(15, 15))
plt.subplot(3,1,2)
plt.semilogy(t[:-1], h_hist, 'o-', color ='firebrick')
plt.title("Step Size h(t)")
plt.xlabel("Time t")
plt.ylabel("Step size (h)")
plt.grid()

#Error estimate vs. time
plt.subplot(3,1,3)
plt.semilogy(t[1:], err_hist, 'o-', color = 'skyblue')
plt.axhline(1e-4, linestyle='--')
plt.title("Error Estimate")
plt.xlabel("Time t")
plt.ylabel("Error estimate")
plt.grid()

#FIXED VS.ADAPTIVE 

fig, ax2 = plt.subplots(figsize=(14,7))

# Adaptive 
ax2.plot(t, y, 'o-', label="Adaptive", color = 'lightgreen')

# Fixed
ax2.plot(t_fixed, y_fixed, '--', label="Fixed (h=0.01)", color= 'b')

ax2.set_xlabel("Time t")
ax2.set_ylabel("Solution y(t)")
ax2.set_title("Adaptive vs Fixed Step Solution")
ax2.grid()
ax2.legend()

plt.show()

#Accepted vs. Rejected steps

fig, ax = plt.subplots(figsize=(6,4))

labels = ["Accepted", "Rejected"]
values = [stats['accepted'], stats['rejected']]

ax.bar(labels, values)
ax.set_title("Accepted vs Rejected Steps")
ax.set_ylabel("Number of Steps")
ax.grid(axis='y')

plt.show()
```
