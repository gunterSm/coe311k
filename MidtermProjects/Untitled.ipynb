 The system we are using is a mechanical system with high damping, specifically,
 an overdamped spring-mass system where the mass is negligible compared to the
#damping and spring forces. A strut for airplane landing gear falls under this
# category, as it tries to minimize large oscillations as quickly as possible.
#
# **The governing ODE**:
#
# dx/dt = -5000x + 0.002sin(t),  with initial condition x(0) = 1 m
#
# This comes from the standard form dx/dt = -(k/c)x + (1/c)F(t), where k/c = 5000 s^-1
# is the ratio of the spring constant to the damping coefficient, and F(t) = 10sin(t) N
# is a slowly varying external force. The 1/c factor scales the force term down to 0.002.
#
# **Physical meaning of variables and parameters**:
#
# x(t) is the displacement of the system in meters, starting at x(0) = 1 m, meaning
# the strut begins compressed by one meter at the moment of contact. The -5000x term
# drives the system rapidly back toward equilibrium; this is the dominant fast dynamics.
# The 0.002sin(t) term represents the slow external forcing. It oscillates on the scale
# of seconds, while the damping acts on the scale of milliseconds. That separation of
# time scales is exactly what makes this equation stiff.
#
# **Why this system is important in practice**:
#
# Getting the damping right matters a lot in practice. If the system is overdamped,
# the strut compresses too slowly, and the full landing load gets transferred to the
# airframe as a hard jolt - at best uncomfortable, at worst structurally damaging.
# If it is underdamped, the strut bounces and the aircraft oscillates on the runway,
# which is a handling and safety problem. The stiffness of this ODE is not just a
# numerical curiosity; it reflects the real physical challenge of a system that snaps
# to equilibrium in milliseconds while being driven by forces that vary over seconds.
